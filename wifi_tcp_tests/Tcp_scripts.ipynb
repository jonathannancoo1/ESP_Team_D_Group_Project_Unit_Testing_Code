```python
import socket

# Server IP and Port
server_ip = '10.225.131.5'
server_port = 8080

# Create a TCP socket
client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

try:
    # Connect to the server
    client_socket.connect((server_ip, server_port))
    print("Connected to server at {}:{}".format(server_ip, server_port))

    # Send the message
    message = "Hello, World!"
    client_socket.sendall(message.encode())
    print("Message sent:", message)

except ConnectionError as e:
    print("Connection error:", e)

```

```python
import socket

def get_ipv4_address():
    # Create a socket to connect to a common server to get the IPv4 address
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
        # Use Google DNS to determine the IPv4 address (doesn't actually send data)
        s.connect(("8.8.8.8", 80))
        ipv4_address = s.getsockname()[0]
    except Exception as e:
        print(f"Error occurred: {e}")
        ipv4_address = None
    finally:
        s.close()
    
    return ipv4_address

# Print your IPv4 address
print("Your IPv4 address is:", get_ipv4_address())
```

```python

# Unless required by applicable law or agreed to in writing, this
# software is distributed on an "AS IS" BASIS, WITHOUT WARRANTIES OR
# CONDITIONS OF ANY KIND, either express or implied.

# -*- coding: utf-8 -*-

import socket
import sys

# -----------  Config  ---------- 
IP_VERSION = 'IPv4'
PORT = 8080
# -------------------------------

if IP_VERSION == 'IPv4':
    family_addr = socket.AF_INET
elif IP_VERSION == 'IPv6':
    family_addr = socket.AF_INET6
else:
    print('IP_VERSION must be IPv4 or IPv6')
    sys.exit(1)


try:
    sock = socket.socket(family_addr, socket.SOCK_STREAM)
except socket.error as msg:
    print('Error: ' + str(msg[0]) + ': ' + msg[1])
    sys.exit(1)

print('Socket created')
    
try:
    sock.bind(('', PORT))
    print('Socket binded')
    sock.listen(1)
    print('Socket listening')
    conn, addr = sock.accept()
    print('Connected by', addr)
except socket.error as msg:
    print('Error: ' + str(msg[0]) + ': ' + msg[1])
    sock.close()
    sys.exit(1)

while True:
    data = conn.recv(128)
    if not data: break
    data = data.decode()
    print('Received data: ' + data)
    reply = 'OK: ' + data
    conn.send(reply.encode())
conn.close()
```
